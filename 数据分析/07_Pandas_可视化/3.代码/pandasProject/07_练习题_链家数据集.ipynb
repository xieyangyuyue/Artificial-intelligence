```python
# 1. 加载数据集.
house_df = pd.read_csv('./data/LJdata.csv')
house_df
```

```python
# 2. 修改列名为 -> 英文.
house_df.columns = ['district', 'address', 'title', 'house_type', 'area', 'price', 'floor', 'build_time', 'direction', 'update_time', 'view_num', 'extra_info', 'link']
house_df
```

- 完成具体的需求

```python
# 1. 找到租金最低，和租金最高的房子
# 思路1: 排序.
house_df.sort_values('price').head(1)       # 最便宜的.
house_df.sort_values('price').tail(1)       # 最贵的.

# 思路2: 筛选
house_df[house_df.price == house_df.price.min()]
house_df[house_df.price == house_df.price.max()]

# 思路3: 最大 或者 最小的 n个
house_df.nlargest(1, 'price')       # 最大的n个, 参1: n个, 参2: 排序字段   可以理解为是: 思路1的简化版
house_df.nsmallest(1, 'price')      # 最小的n个
```

```python
# 2. 找到最近新上的10套房源
house_df.sort_values('update_time', ascending=False).head(10)
```

```python
# 3. 查看所有更新时间
house_df.update_time.unique()
```

```python
# 4. 查看看房人数的平均值, 最大值, 最小值
house_df.view_num.mean()
house_df.view_num.max()
house_df.view_num.min()

house_df['view_num'].mean()
house_df['view_num'].max()
house_df['view_num'].min()

house_df[['view_num']].describe()       # DataFrame类型
house_df['view_num'].describe()         # Series类型
```

```python
# 5. 查看不同看房人数的房源数量，as_index = False 分组字段不作为行索引（默认为True)
# 思路: 基于看房人数分组, 统计房源数量
house_df.groupby('view_num').address.count()
house_df.groupby('view_num')['address'].count()
house_df.groupby('view_num').agg({'address':'count'})

# 完整版
tmp_df = house_df.groupby('view_num', as_index=False).agg({'address':'count'})
# 修改列名
tmp_df.columns = ['view_num', 'house_count']
tmp_df
```

```python
# 6. 查看房租价格的分布, 例如: 平均值, 标准差, 中位数...
house_df.price.describe()
house_df.price.mean()
house_df.price.max()
```

```python
# 7. 找到看房人数最多的朝向
# 思路: 基于朝向分组, 统计 看房人数 总和
tmp_df = house_df.groupby('direction', as_index=False).agg({'view_num':'sum'})
# 写法1: 排序
tmp_df.sort_values('view_num', ascending=False).head(1)
# 写法2: nlargest()函数 
tmp_df.nlargest(1, 'view_num')
```

```python
# 8. 查找最受欢迎的房型
# 思路: 基于房型分组, 统计 看房人数 总和
tmp_df = house_df.groupby('house_type', as_index=False).agg({'view_num':'sum'})
# 写法1: 排序
tmp_df.sort_values('view_num', ascending=False).head(1)
# 写法2: nlargest()函数 
tmp_df.nlargest(1, 'view_num')
```

```python
# 9. 查找房子的平均租房价格 （元/平米）
# 思路1: 计算每套房子的平均价格, 然后在统计所有平均价格的 平均值.
house_df['price_area_m²'] = house_df.price / house_df.area
house_df['price_area_m²'].mean()            # 87.72268429900454

# 思路2: 总租金价格 / 总面积.
house_df.price.sum() / house_df.area.sum()  # 86.36679645198356


# 已知: A房子 -> 100平, 租金: 20元,    B房子 -> 50平, 租金: 100元
# 思路1: 所有房子均价的平均值.
# (20 / 100) + (100 / 50) = 0.2 + 2 = 2.2  ->  2.2 / 2 = 1.1元/m²

# 思路2: 总租金价格 / 总面积.
# (20 + 100) / (100 + 50) = 120 / 150 = 0.8元/m²
```

```python
# 10. 找到出租房源最多的小区
# 思路: 基于 地址(小区)分组, 统计 房源数量, 获取最大值.
house_df[['address', 'view_num']].groupby('address', as_index=False).count()
house_df.groupby('address', as_index=False).view_num.count()        # 效果同上. 

tmp_df = house_df[['address', 'view_num']].groupby('address', as_index=False).agg({'view_num':'count'})
tmp_df.columns = ['address', 'house_count']
# 思路1: 排序 + head()
tmp_df.sort_values('house_count', ascending=False).head(1)

# 思路2: nlargest()函数
tmp_df.nlargest(1, 'house_count')
```
